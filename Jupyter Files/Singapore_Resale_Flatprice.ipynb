# Handling the Outliers through IQR method

```python
sns.boxplot(df,orient='h')
```

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
    lr,ur=remove_outlier(df["resale_price"])
    df["resale_price"]=np.where(df["resale_price"]> ur,ur,df["resale_price"])
    df["resale_price"]=np.where(df["resale_price"]<lr,lr,df["resale_price"])
```
