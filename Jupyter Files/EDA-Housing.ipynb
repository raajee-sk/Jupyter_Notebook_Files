```python
from scipy import stats
```

```python
z=np.abs(stats.zscore(housing_df))
```

```python
print(z)
```

```python
threshold=3
np.where(z>3)
```

```python
q1=housing_df.quantile(0.25)
q3=housing_df.quantile(0.75)
iqr=q3-q1
iqr
```

```python
np.where((housing_df<(q1-1.5*iqr))|(housing_df>(q3+1.5*iqr)))
```

```python
housing_df1=housing_df[(z<3).all(axis=1)]
```

```python
housing_df1.shape,housing_df.shape
```

```python
housing_df2=housing_df.copy()
```

```python
for i, j in zip(np.where(z > threshold)[0], np.where(z > threshold)[1]):# iterate using 2 variables.i for rows and j for columns
    housing_df2.iloc[i,j] = housing_df.iloc[:,j].median()
```

```python
z = np.abs(stats.zscore(housing_df2))
np.where(z > threshold)
```

```python
housing_df2.shape,housing_df.shape
```

```python
housing_df_out = housing_df[~((housing_df < (q1 - 1.5 * iqr)) |(housing_df > (q3 + 1.5 * iqr))).any(axis=1)] # rows without outliers
housing_df_out.shape
```

```python
housing_df3=housing_df.copy()
```

```python
for i, j in zip(np.where(housing_df3 < q1 - 1.5 * iqr)[0], np.where(housing_df3 < q1 - 1.5 * iqr)[1]): 
    
    whisker  = q1 - 1.5 * iqr
    housing_df3.iloc[i,j] = whisker[j]
    
```

```python
for i, j in zip(np.where(housing_df3 > q3 + 1.5 * iqr)[0], np.where(housing_df > q3 + 1.5 * iqr)[1]):
    
    whisker  = q3 + 1.5 * iqr
    housing_df3.iloc[i,j] = whisker[j]
    
```
