```python
#Four columns =[Ecom,SalesFImage,OrdBilling,DelSpeed] seems to have outliers.We are going to treat them  through IQR Method.
```

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
Outlier=["Ecom","SalesFImage","OrdBilling","DelSpeed"]
```

```python
for i in Outlier:
    lr,ur=remove_outlier(df[i])
    df[i]=np.where(df[i]> ur,ur,df[i])
    df[i]=np.where(df[i]<lr,lr,df[i])
    
```
