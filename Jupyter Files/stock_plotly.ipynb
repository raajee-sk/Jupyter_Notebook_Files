```python
import seaborn as sns
```

```python
import plotly.express as px
```

```python
import matplotlib.pyplot as plt
```

```python
df_stocks=px.data.stocks()
```

```python
df_stocks
```

```python
plt.figure(figsize=(12,8))
x=df_stocks["date"]
y=df_stocks["GOOG"]
plt.xlabel("Date")
plt.ylabel("Google stocks")
plt.title("Google Stock by date",color='blue')

plt.plot(x,y,marker="o",linestyle="-",color='r')
```

```python
df=sns.load_dataset("tips")
```

```python
df
```

```python
df["day"].value_counts().sort_values(ascending=False).plot(kind='bar')
```

```python
plt.pie(df["day"].value_counts(),autopct='%1f%%',radius=1.5,labels=['Sat','Sun','Thur','Fri'])
circle=plt.Circle((0,0),radius=0.5,color='white')
plot=plt.gcf()
plot.gca().add_artist(circle)
```

```python
plt.hist(df.total_bill,bins=20,color='green')
plt.title("Total bill Histogram")
plt.xlabel("Bins")
plt.ylabel("Total_bill")
plt.text(x=32.5,y=35,s="Total_bill average:")
plt.text(x=46.5,y=35,s=round(df.total_bill.mean(),2))
plt.text(x=32.5,y=33,s="Total_bill median:")
plt.text(x=46.5,y=33,s=round(df.total_bill.median(),2))
plt.text(x=32.5,y=31,s="Total_bill mode:")
plt.text(x=46.5,y=31,s=df.total_bill.mode()[0])
```
