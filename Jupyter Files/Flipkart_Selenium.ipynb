```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
import time
```

```python
driver=webdriver.Chrome()
driver.maximize_window()
driver.get("https://www.flipkart.com/footwear/pr?sid=osp&marketplace=FLIPKART")
```

```python
details={'Brands':[],'Product_Description':[],
        'Discount_Price':[],'Actual_Price':[],
        'Discount_Percent':[],"Links":[],'Image_Link':[]}
```

```python
for i in range(5):
    Elementboxes=driver.find_elements(By.CSS_SELECTOR,"div._1xHGtK")
    for i in Elementboxes:
        try:ImagesElements=i.find_element(By.CSS_SELECTOR,'img._2r_T1I').get_attribute('src')
        except:ImagesElements=None

        try:BrandsElements=i.find_element(By.CSS_SELECTOR,'div._2WkVRV').text
        except:BrandsElements=None 

        try:ProductdescriptionElements=i.find_element(By.CSS_SELECTOR,'a.IRpwTa').get_attribute("title")
        except:ProductdescriptionElements=None

        try:DiscountpriceElements=i.find_element(By.CSS_SELECTOR,"div._30jeq3").text
        except:DiscountpriceElements=None 

        try:ActualpriceElements=i.find_element(By.CSS_SELECTOR,"div._3I9_wc").text
        except:ActualpriceElements=None

        try:DiscountpercentElements=i.find_element(By.CSS_SELECTOR,"div._3Ay6Sb").text
        except:DiscountpercentElements=None 

        try: LinksElement=i.find_element(By.CSS_SELECTOR,'a.IRpwTa').get_attribute('href') 
        except:LinksElement=None  

        details['Image_Link'].append(ImagesElements)
        details['Brands'].append(BrandsElements)
        details['Product_Description'].append(ProductdescriptionElements)
        details['Discount_Price'].append(DiscountpriceElements)
        details['Actual_Price'].append(ActualpriceElements)
        details['Discount_Percent'].append(DiscountpercentElements)
        details['Links'].append(LinksElement)    
    driver.find_element(By.XPATH,"//span[contains(text(),'Next')]").click()
    time.sleep(3)     
```

```python
for i in details.keys():
    print(i,len(details[i]))
```

```python
df=pd.DataFrame(details)
```
